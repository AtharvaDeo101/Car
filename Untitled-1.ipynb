## Data Cleaning


```python
#we are going to convert the float data into interger data

df[['wheelbase','carlength','carwidth','carheight','boreratio','stroke','compressionratio','price']]  = df[['wheelbase','carlength','carwidth','carheight','boreratio','stroke','compressionratio','price']].astype(int)
```

```python
df = df.drop('symboling', axis = 1)
```

```python
df
```

```python
df['CarName'] = df['CarName'].str.split(' ').str[0]
```

```python
#Performing One Hot encoding on categorical data
data = ['CarName','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem']
df_encoded = pd.get_dummies(df, columns= data, drop_first= True)
df_encoded = df_encoded.replace({True: 1, False: 0})
```

```python
df_encoded
```

```python
df_encoded.describe()
```

```python
df_encoded.to_csv('cleanedCardata.csv')
```

## Traning Model

```python
x = df_encoded.drop('price', axis = 1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.2,random_state= 2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
lr = LinearRegression()
rfr = RandomForestRegressor()
```

```python
#performing hyper-parameter tuning using GridSearchCV - we could also find paremeters using cross fold verification but we have to use loop which reduces it efficiency
#for small dataset we can use gridSearchCV while for BIG dataset we have to use RandomearchCV

from sklearn.model_selection import GridSearchCV
```

```python

clf = GridSearchCV(estimator=lr,param_grid= {
    'fit_intercept' : [True, False],
    'copy_X': [True,False],
    'positive': [True,False]
},  cv= 5)
```

```python
clf.fit(x_train,y_train)
```

```python
# Check if there are any non-numeric columns in X_train
print(x_train.dtypes)
```

```python
clf.cv_results_ 
```

```python
new_df = pd.DataFrame(clf.cv_results_)
new_df
```

```python
clf.best_params_
```

```python
lr  = LinearRegression(copy_X= True, fit_intercept= False, positive= False)
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("r2 score : ",r2_score(y_test,y_pred))
print("mean_squared_error", mean_squared_error(y_test,y_pred))
```
