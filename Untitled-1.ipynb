## Data Cleaning


```python
#we are going to convert the float data into interger data

df[['wheelbase','carlength','carwidth','carheight','boreratio','stroke','compressionratio','price']]  = df[['wheelbase','carlength','carwidth','carheight','boreratio','stroke','compressionratio','price']].astype(int)
```

```python
#Performing One Hot encoding on categorical data
data = ['fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem']
df_encoded = pd.get_dummies(df, columns= data, drop_first= True)
df_encoded = df_encoded.replace({True: 1, False: 0})
```

```python
df_encoded.drop('symboling', axis=1, inplace=True)
```
